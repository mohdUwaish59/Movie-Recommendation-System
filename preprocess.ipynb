```python
movies = pd.read_csv('Dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('Dataset/tmdb_5000_credits.csv')
```

```python
movies_tags = movies_merged[['id','title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
avatar = movies_tags.iloc[0]['crew']

avatar
```

```python
import ast
def filter_genres(movie_genres):
    genres = []
    for i in ast.literal_eval(movie_genres):
        genres.append(i['name'])
    return genres

movies_tags['genres'] = movies_tags['genres'].apply(filter_genres)
movies_tags['keywords'] = movies_tags['keywords'].apply(filter_genres)
```

```python
def movie_cast(movies_tags):
    cast = []
    counter = 0
    for i in ast.literal_eval(movies_tags):
        if counter != 3:
            cast.append(i['name'])
        else:
            break
    return cast
movies_tags['cast'] = movies_tags['cast'].apply(filter_genres)
```

```python
def movie_director(movies_tags):
    director = []
    for i in ast.literal_eval(movies_tags):
        if i['job'] == 'Director':
            director.append(i['name'])
            break
    return director
movies_tags['crew'] = movies_tags['crew'].apply(movie_director)
```

```python
movies_tags['overview'] = movies_tags['overview'].str.split()
```

```python
movies_tags
```

```python
movies_tags['genres'] = movies_tags['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_tags['keywords'] = movies_tags['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_tags['crew'] = movies_tags['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies_tags['cast'] = movies_tags['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_tags['tags'] = movies_tags['overview'] + movies_tags['keywords'] + movies_tags['genres'] + movies_tags['cast'] + movies_tags['crew']
```

```python
movies_tags
```

```python
movies_df = movies_tags[['id', 'title', 'tags']]
movies_df
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x: " ".join(x) if isinstance(x, list) else x)
```

```python
movies_df['tags'] = movies_df['tags'].str.lower()
```

```python
movies_df
```

```python
movies_df.dropna(inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=5000, stop_words='english')
vectors = vectorizer.fit_transform(movies_df['tags']).toarray()
```
